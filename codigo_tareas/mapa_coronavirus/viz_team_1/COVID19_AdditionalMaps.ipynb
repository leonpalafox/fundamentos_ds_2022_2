```python
import pandas as pd
# lectura de datos 
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
columns =[]
today = df.columns[-1]
columns.append(today) # obtenemos la informacion mas actual
[columns.append(x) for x in df.columns[ 0:4 ] ]    # agregamos las columnas de nombres pais-ciudad-coordenadas
columns
```

```python
df.head()
```

```python
df_map = df[columns]
df_map.head()
```

```python
info_map = df_map.groupby('Country/Region').sum().sort_values([today],ascending=False)
```

```python
import folium
from folium import plugins
import numpy as np
#df_map = df_map.dropna()
#df_map = df_map.reset_index(drop=True) # hay ciudad sin nombres, mejor no las visualizamos
```

```python
mapit = folium.Map( location = [28.6631453,-53.0675644], zoom_start = 2)
marker_cluster_ = folium.plugins.MarkerCluster()
for idx in range(0,len(df_map) ):
    marker_ = folium.Marker( location = df_map[  ['Lat', 'Long'] ].iloc[idx], 
                            popup = str(df_map['Province/State'][ idx ]) + ":" + str(df_map[today][ idx ]) ,  
                           icon = folium.Icon(color='red') )
    marker_cluster_.add_child(marker_)
mapit.add_child(marker_cluster_)
```

```python
# ahora con burbujas 
############
m = folium.Map(
    location = [28.6631453,-53.0675644],
    tiles='Stamen Toner',
    zoom_start=2
)
for i in range(0, len(info_map)):
    folium.CircleMarker(
    radius =  ( np.log(info_map[today][i]) )*2,
    location = [info_map.iloc[i]['Lat'], info_map.iloc[i]['Long']],
    popup = info_map.index[i],
    color='#3186cc',
    fill=True,
    fill_color='crimson', 
    fill_opacity = .5
).add_to(m)
m
```
